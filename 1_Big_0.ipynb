# BIG-O Baby!

Let's start by coding two different implementations of a summing function: 

## Implementation #1

# $$\sum_{i=0}^{n=5}i = 15$$

```python
def sum1(n):
    final_sum = 0
    for x in range(n+1):
        final_sum += x
        
    return final_sum
```

```python
sum1(5)
```

## Implementation #2

# $$\sum_{i=0}^{n=5}i = \frac{n(n+1)}{2} = 15$$

```python
def sum2(n):
    return (n*(n+1))/2
```

```python
sum2(5)
```

We can compare both implementations in two ways:

1. The amount of space they take in memory.
2. The amount of time they take to run.

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

Implementation two runs much faster than the first, but measuring in time is not hardware independent. 
Big-O notation solves that problem for us.

## $O(n)$: How quickly runtime will grow relative to the input as the input gets arbitrarily large.

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Set up runtime comparisons
n = np.linspace(1,10,1000)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot setup
plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```

## $1$ = Constant

## $log(n)$ = Logarithmic

## $n$ = Linear

## $nlog(n)$ = Log Linear

## $n^2$ = Quadratic

## $n^3$ = Cubic

## $2^n$ = Exponential

## $1$ = Constant
Regardless of the input size of the function, it will always take
a constant step size.

```python
lst = [1,2,3,4,5]

def func_constant(n):
    print(n[0])


func_constant(lst)
    
```

## Dropping Constants
# $O(2n) \equiv O(n)$

```python
def drop_constants(n):
    
    for x in n:
        print(x)
    for x in n:
        print(x)
        
drop_constants(lst)
```

## $n$ = Linear
The number of print statements scale linearly with $n$.

```python
def func_linear(n):
    for x in n:
        print(x)
        
func_linear(lst)
    
```

## $n^2$ = Quadratic

```python
def func_quadratic(n):
    for x in n:
        for y in n:
            print(x, y, end=" ")
func_quadratic(lst)
    
```

## Best Case vs. Worst Case 

```python
print(lst)
def matcher(match, lst):
    for x in lst:
        if x == match:
            return True
    return False
```

```python
matcher(1,lst) #O(1)
```

```python
matcher(6,lst) #O(n)
```

## Space Complexity


```python
def create_list(n):
    new_list = []
    for num in range(n):
        new_list.append('new')
    print(new_list)
```

```python
create_list(5)
```

The size of the new list created scales linearly with the value of n.

```python
def printer(n):
    for x in range(n):
        print('hello')
```

```python
printer(5)
```

Time Complexity: $O(n)$ , Space Complexity: $O(1)$

## Comparing Python List Methods

### Appending to the end of a list

```python


def method_1():
    l = []
    for n in range(1000):
        l = l + [n]
method_1()
    
```

### Concatenating a list

```python
def method_2():
    l = []
    for n in range(1000):
        l.append(n)
method_2()
```

### List Comprehension

```python
def method_3():
    l = [n for n in range(1000)]
method_3()
```

### Using the built-in range function 

```python
def method_4():
    list(range(1000))
method_4()
```

```python
%timeit method_1()
%timeit method_2()
%timeit method_3()
%timeit method_4()
```

So the range function wins, with the list comprehension coming in a close second! If you append to the end of a list, you don't get the job.

## Comparing Python Dictionary Methods
